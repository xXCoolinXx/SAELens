```python
from pathlib import Path
import ipywidgets as widgets
from IPython.display import display, HTML

plot_paths = sorted(
    Path("./7ajlxjnb_100003840").glob("*.html")
)  # folder with your HTML plots
idx = 0
out = widgets.Output()


def render_plot(i):
    path = plot_paths[i]
    with out:
        out.clear_output(wait=True)
        html = path.read_text(encoding="utf-8")
        display(HTML(html))  # embeds the Plotly HTML directly


def on_next(_):
    global idx
    idx = (idx + 1) % len(plot_paths)
    render_plot(idx)


def on_prev(_):
    global idx
    idx = (idx - 1) % len(plot_paths)
    render_plot(idx)


prev_btn = widgets.Button(description="◀ Prev")
next_btn = widgets.Button(description="Next ▶")
prev_btn.on_click(on_prev)
next_btn.on_click(on_next)

display(widgets.HBox([prev_btn, next_btn]))
display(out)
render_plot(idx)
```
